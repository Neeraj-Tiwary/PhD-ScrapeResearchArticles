```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options as EdgeOptions
import time


options = EdgeOptions()
#options.add_argument("--headless=new")
#options.add_argument("--disable-extensions")
#options.add_argument("--log-level=3")
options.page_load_strategy = 'normal'

driver = webdriver.Edge(options=options)

#driver = webdriver.Edge()
#driver = webdriver.Chrome()
driver.get(url)
time.sleep(3)

try:
    # element = WebDriverWait(driver, 10).until(
    #     EC.presence_of_element_located((By.TAG_NAME, "xpl-results-item"))
    # )
    
    contents = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CLASS_NAME, "result-item-content"))
    )
    
    # contents = driver.find_elements(by=By.TAG_NAME, value='xpl-results-item')
    print(len(contents))
except:
    driver.quit()
```

```python
for content in contents:
    print (content.text)
    print("-"*65)
```

```python
headers = [content.find_element(By.TAG_NAME,'a') for content in contents]
print(len(headers))
for header in headers[:5]:
    print(header.text)
```

```python
authors = [content.find_element(By.TAG_NAME,'xpl-authors-name-list') for content in contents]
print(len(authors))
for author in authors[:5]:
    print(author.text)
```

```python
highlights = [content.find_element(By.CLASS_NAME,'description.text-base-md-lh') for content in contents]
titles = [highlight.find_element(By.TAG_NAME,'a') for highlight in highlights]
print(len(titles))
for title in titles[:5]:
    print(title.text)

# print(len(highlights))
# for highlight in highlights[:10]:
#     print(highlight.text)
```

```python
publisher_info = [content.find_element(By.CLASS_NAME,'publisher-info-container') for content in contents]
years = [year.find_element(By.TAG_NAME,'span') for year in publisher_info]
years = [year.text.split()[1] for year in years]

print(len(years))
for year in years[:5]:
    print(year)

```

```python
titles_list = []
authors_list = []
publication_list = []
years_list = []

for title in headers:
    titles_list.append(title.text)
    
for author in authors:
    authors_list.append(author.text)
    
for publication in titles:
    publication_list.append(publication.text)
    
for year in years:
    years_list.append(year)
```

```python
import pandas as pd

df = pd.DataFrame(zip(titles_list, authors_list, publication_list, years_list), columns=['Title', 'Authors', 'Publication Title', 'Year'])
df
```
